## LOAD CSV FILE INTO PANDAS

```python
import pandas as pd
import numpy as np

student = pd.read_csv('https://raw.githubusercontent.com/davay521/IS362_PROJECT2/master/student_data.csv')
student
```

# Sorted the mess a bit to get a clearer picture of the data


```python
studentSort = student.sort_values(['Country','Region','Gender'],ascending = [True,True,False])

studentSort
```

## Is lefthanded-ness more common in male students or female students?


```python
student.groupby('Gender')['Handed'].value_counts()
```

## What country's students have The longest commute to school?

```python
student.groupby('Country')['Travel_time_to_School'].mean().sort_values(ascending=False)
```

## What method of travel is used most in each country

```python
student.groupby('Country')['Travel_to_School'].value_counts().sort_values(ascending= False)
```
